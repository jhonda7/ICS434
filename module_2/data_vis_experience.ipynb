```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### First I will set the seed to 42 as stated in the requirements. Then I will use the NumPy library to generate a NumPy array of length 1000 containing random selections from the list: ["CS", "Engineering", "Math", "Physics"]. This is dont using NumPy's random choice method.

```python
# set the seed to 42 (requirement)
np.random.seed(42)

# generate the
students_affiliations = np.random.choice(["CS", "Engineering", "Math", "Physics"], size=1000, replace=True)
```

##### Now, I will convert the students_affiliations np.array into a regular Python list so that I can use the '.count()' method (discussed shortly below). I also create a list containing the affiliations that I will use as the labels for my x axis in my plot. Then I create a list containing the counts of each of the affiliations within students_affiliations.

##### NOTE: I originally generated students_affiliations using NumPy because this was the only way to get numerical results congruent to the example above. I tried to use the 'random' package via 'import random' - but this yielded different numerical values from the example shown above. I also needed to use the '.count()' method - hence why I create students_affiliations as a NumPy array then transform it into a list.

```python
students_affiliations = students_affiliations.tolist()
categories = ["Physics", "CS", "Math", "Engineering"]
values = [students_affiliations.count("Physics"), students_affiliations.count("CS"), students_affiliations.count("Math"), students_affiliations.count("Engineering")]
```

##### Below I create the plot using matplotlib's bar plot. I abide by the required color and alpha parameters

```python
plt.figure(figsize=(10,6))
plt.xlabel("Disciplines", fontsize = 15)
plt.ylabel("Counts", fontsize = 15)
plt.title("Distributions of Students in the Program", color = 'green', fontsize = 25)
plt.bar(categories, values, color = "#FF5733", alpha = 0.55)
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

##### I use NumPy's random number generator to create two arrays of values: hours_studied and exam_score. Then I use matplolib's scatter plot function. to change the symbols from circles to squares, I adjust the 'marker' parameter and set it to 's'.

```python
hours_studied = np.random.randint(0, 24, 200)
exam_score = np.random.randint(0, 100, 200)

plt.figure(figsize=(12,6))
plt.scatter(hours_studied, exam_score, color = "#FF5733", marker = 's')

plt.xlabel("Hours of Study", fontsize = 15)
plt.ylabel("Exam Score", fontsize = 15)
plt.title("Representation of Hours Studied and Exam Score", color = "green", fontsize = 25)
```
